```python
# Import necessary packages 
import glob 
import ast

import pandas as pd

import numpy as np
from scipy.optimize import curve_fit
from sklearn.metrics import r2_score

import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("whitegrid")
```

We will  now define the functions needed for the data extraction and exploratory plots.

```python
#########   Function to  extract data correscponding to one cycle   ######

def extract_from_file(data,n_wells=72):
    
    """A function that extracts fluorescence data from one pcr cycle"""
    
    cycle=[]
    
    # iterrate over all floorescence values measured for all the wells for each pcr cycle
    for i in range(n_wells):
        cycle.append(data[i][2])# append fluorescence values to a cycle list 
        
    return(cycle)
```

```python
#####  Function that returns a data structure created out of all the data from an experiment #####

def extract_data(path):
    
    """A function that extracts fluorescence data from all the files from an experiment
     and returns a resulting data structure in the form of list of lists"""
    
    files = glob.glob(path) 
    data_structure=[]

    # Read and extract data from all text files in the directory. The fluorescence data from all the files will be
    #written in the data_structure list of lists
    for name in files:
        with open(name) as f:
    
            data = f.read()
            data=ast.literal_eval(data)
            cycle=(extract_from_file(data))
            data_structure.append(cycle)       
            
    return(data_structure)        
                                         
```

```python
#########   Function to  represent all time series in one plot   ######

def plot_fluorescence(df,n_wells=72):
    """A function that plots all fluorescence data in one plot"""
    
    fig, ax=plt.subplots(figsize=(10,7))
    
    for i in range(n_wells):
        ax.plot(df.index,df.iloc[:,i+1],label=df.columns[i+1])
    
   
    plt.ylabel('Fluorescence',fontsize=16)
    plt.xlabel('Cycle number',fontsize=16)
    plt.title('Fluorescence over a time course during a PCR',fontsize=18)
    plt.legend(ncol=3,bbox_to_anchor=(0.,-0.100, 1.55, 1.15), loc=1,prop={'size': 12})
    plt.show()  
```

```python
#########   Function to split time series in 8 subplotsplot   ######


def subplots_fluorescence(df):
    
     """A function that divides fluorescence data in 8 subplots"""
     
     fig, ax=plt.subplots(4,2, figsize=(25,20))
    
     n=1
    
     for row in range(4):
        for col in range(2):
            
            for i in range(9):
                
                ax[row,col].plot(df.index,df.iloc[:,n+i],label=df.columns[n+i])
        
            n=n+9
            ax[row,col].legend(frameon=False,loc='upper left',fontsize=14)
            ax[row,col].set_xlim([-7, 41])
            ax[row,col].set_ylabel('Fluorescence',fontsize=18)
            ax[row,col].set_xlabel('Cycles',fontsize=18)
            ax[row,col].tick_params(labelsize=15)
            
                            
        plt.suptitle('Fluorescence over a time course during a PCR',fontsize=22,y=0.91)
     plt.show()   
```

## Fitting data with a sigmoid-like model

We can see that the fluorescence signals of amplified samples follow a sigmoid-like shape, so in the next step of the analysis, we will select the data that putatively indicates DNA amplification 
and we will perform least-squares fits of  a sigmoid-like model to this data.

Note that after the PCR cycle number 34, we can observe a sudden drop in fluorescence intensity in all the time series (is it due to the dye bleaching?) till the level of their corresponding baselines.
We need to take this phenomenon into account and fit the time series data till the cycle number 34. 

Also, ideally, before performing fits, we would correct the drift in the  tada (as many traces of the fluorescence time series seem to have negative drift), 
however, we will skip this step here for the sake of saving time. You can see an example of how the data drift can be corrected here: https://github.com/katerynalemishko/protein-unfolding/blob/main/data-preparation-magnetic-tweezers.ipynb
        
Let's start with defing the fuction to fit the data.        

**Sigmoid function:**

```python
from IPython.display import display, Math, Latex
display(Math(r'f(x) = \frac{L} {1 + e^{-k(x-x0)}}+b'))
```

Here, L is responsible for scaling the output range from [0,1] to [0,L], L models the fluorescence amplitude;

b adds bias to the output and changes its range from [0,L] to [b,L+b], b models the signal base line;

k controls the shape of the curve, the larger the value of k, the later the fast growth of a sigmoid starts, and the steeper the growth;

x0 is the point in the middle of the Sigmoid, i.e. the point where Sigmoid should originally output the value 1/2 [since if x=x0, we get 1/(1+exp(0)) = 1/2].

```python
####        Define sigmoid function to use for the fits   #######

def sigmoid(x, L ,x0, k, b):
    y = L / (1 + np.exp(-k*(x-x0))) + b 
    return (y) 
```

Now, let's define two functions: one of the functions will perform fits of the sigmoid model to the selected data and return a data frame containing
obtained optimized parameters L,x0,k and b together with the r_squared, and the other function will plot the fluorescence data together with the simulated fits.

Both functions will take as an input a data frame containing all the fluorescence data from a particular experiment, e.g. data frame 
**fluorescence_data_exp_1** and a list with the names of the selected wells that we want to analyse.

```python
#######   Function Sigmoid fit : returns a dataframe with the optimized parameters and the r_squared of the fit##########

def sigmoid_fit(df,wells_to_fit):
    """A function that fits sigmoid model to selected fluorescence data and returns a dataframe with the
    optimized parameters and the r_squared of the fit """

    L_param=[]
    x0_param=[]
    k_param=[]
    b_param=[]
    r_squared=[]
    
    for i in wells_to_fit:
        xdata, ydata = df.index[df.index<34], df[df.index<34][i]
    
        p0 = [max(ydata), np.median(xdata),1,min(ydata)] 
        popt, pcov = curve_fit(sigmoid, xdata, ydata,p0, method='dogbox')
        L ,x0, k, b =popt  
    
        ypred=sigmoid(xdata,*popt)
        r_squared.append(r2_score(ydata,ypred))
    
        L_param.append(round(popt[0],3))
        x0_param.append(round(popt[1],3))
        k_param.append(round(popt[2],3))
        b_param.append(round(popt[3],3))
   
    fit_parameters={'L':L_param,'x0':x0_param,'k':k_param,'b':b_param,'well_number':wells_to_fit,'r_squared':r_squared}

    fit_parameters_df=pd.DataFrame(fit_parameters)
   
    return(fit_parameters_df)
```

```python
#######Function that plots sigmoid fits to selected data ##########

def sigmoid_fit_plot(df,wells_to_fit):
    
    """A function that fits sigmoid model to selected fluorescence data and plots the obtainded fits"""
    
    for i in wells_to_fit:
        xdata, ydata = df.index[df.index<34], df[df.index<34][i]
    
        p0 = [max(ydata), np.median(xdata),1,min(ydata)] 
        popt, pcov = curve_fit(sigmoid, xdata, ydata,p0, method='dogbox')
        L ,x0, k, b =popt  
    
        #plot data
        plt.scatter(xdata, ydata,s=15,label=i)
    
        #plot fits
        x_line = np.arange(min(xdata), max(xdata), 1)
        y_line = sigmoid(x_line,L ,x0, k, b)
        plt.plot(x_line, y_line, '--')

    plt.title('Frluorescence data with sigmoid fits',fontsize=16)
    plt.xlabel('Cycle',fontsize=14)
    plt.ylabel('Fluorescence',fontsize=14)
    plt.legend(ncol=3,frameon=False,loc='upper left')
    plt.show() 
```

Let's visualise the fit parameters. To do this, we can write a function so we can reuse it to do the same analysis of data from another experiments.

```python
#########   Function to visualyse compare fit parameters among different wells  ######

def fit_param_visualise(df):
    
     """A function that visualyses fit parameters among different wells """
     
     # Set lists containing the names of the parameters,titles for the sublots and palette names 
     parameters=['L','k','x0','r_squared']
     titles=['Parameter L','Parameter k','Parameter x0','R squared']
     palette=["Blues_d",'Purples','Oranges','Greens']
    
    
     fig, ax=plt.subplots(2,2, figsize=(18,10))
     
     #initializing the subplot index
     n=0
        
     for row in range(2):
        for col in range(2):
            sns.barplot(ax=ax[row,col],x='well_number', y=parameters[n], data=df,palette=palette[n],
                 order=df.sort_values(parameters[n]).well_number)
            ax[row,col].set_ylabel(parameters[n],fontsize=18)
            ax[row,col].set_xlabel('Wells',fontsize=18)
            ax[row,col].tick_params(axis='x', labelrotation=45,labelsize=15)
            ax[row,col].tick_params(axis='y', labelsize=15)
            ax[row,col].set_title(titles[n],fontsize=20)
            
            # index for the next subplot
            n+=1
    
     plt.suptitle('Optimized fit parameters for the selected data',y=0.98,fontsize=25)
     plt.tight_layout()
     plt.show()
```

```python
fit_param_visualise(fit_exp1_params)
```

Now, we can compare the fit parameters for the samples from different wells.
Ideally, we will normalize the fluorescence intensity data so all the fluorescence traces have a baseline around Fluoresccence=0;
in this case, the parameters L (amplitude) and b (baseline) will be also modified to corresponding normalised parametrs. However, for the sake of just showing an example of the analysis, we will use non-normalised parametrs.


You can see that the fluorescence amplitude (parameter L) varies significantly among diffrent wells (from 73000 to 352000). 
We suppose that there should be a criterion according to which a sample can be considered amplified when the fluorecsence time series has a sigmoid-like shape.
According to such criterion, some wells (like well_38 and well_28) might be considered as not amplified.

As it can be seen on the graph, the parameter r_squared that indicates the goodness of the fit varies significantly.
We can clearly see that we obtained six very good fits with r_squared>0.96 (with the maximum r_squared=0.996) corresponding to the data from the wells number 65,9,44,53,11 and 23.
However, other fits are not so good: there were three fits with r_squared<0.8 (for the data corresponding to the wells number 22,21 and 32) and 11 out of 17 fits had r_squared less than 0.9.
We can improve the R squared of the fits if we include an additional parameter that accounts for irregularities in data (occasional points that lie way above or below the fluorescence baseline).    
    
As you see, the values of the parameter k defining the shape of the sigmoid vary significantly among different samples, however the values of the parameter
x0 which defines the number of cycle at which the middle of a sigmoid is reached are very similar for all selected wells (x0 values lie between the cycles 29 and 31).
We might also apply a criterion to classify a sample as amplified or not based on the estimation of the parameters k and x0.

In summary, to classify data as sigmoid-like, the criteria could be as folows:

- The fluorescence intensity amplitude should exceed a certain treshold: **L > L<sub>threshold</sub>**


- Plus, for the sample to be classified as amplified, the parameters k and  x0 
should lie within appropreate ranges so: **k<sub>lower</sub> < k < k<sub>upper</sub>** and  **x0<sub>lower</sub> < x0 < x0<sub>upper</sub>**

- Finaly the R squared value which reflects the goodness of fit should be high enough: **R<sup>2</sup> > R<sup>2</sup><sub>threshold</sub>**
